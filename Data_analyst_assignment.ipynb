## 2. Given an unsorted array of integers, find the length of the longest continuous increasing subsequence (subarray). 

```python
def longest_increasing_subsequence(nums):
    if not nums:
        return 0
    
    max_length = 1
    current_length = 1
    
    for i in range(1, len(nums)):
        if nums[i] > nums[i - 1]:
            current_length += 1
            max_length = max(max_length, current_length)
        else:
            current_length = 1
    
    return max_length

# Test cases
print(longest_increasing_subsequence([1, 3, 5, 4, 7]))  
print(longest_increasing_subsequence([2, 2, 2, 2, 2]))  
```

## 3. Given a list of non negative integers, arrange them such that they form the largest number.

```python
from functools import cmp_to_key

def largest_number(nums):
    def compare(x, y):
        return int(y + x) - int(x + y)

    nums = map(str, nums)
    nums = sorted(nums, key=cmp_to_key(compare))
    return ''.join(nums).lstrip('0') or '0'


print(largest_number([10, 2]))  
print(largest_number([3, 30, 34, 5, 9]))  
```

## 4. Store all the "servlet-name", and "servlet-class" to a csv file from the attached sample_json.json file using Python.

```python
import json
import csv

# Read the JSON file
with open('DT A1 sample_json (1) (1).json', 'r') as json_file:
    data = json.load(json_file)

servlet_data = [(entry['servlet-name'], entry['servlet-class']) for entry in data['web-app']['servlet']]

with open('servlet_data.csv', 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['servlet-name', 'servlet-class'])  
    writer.writerows(servlet_data)
```
